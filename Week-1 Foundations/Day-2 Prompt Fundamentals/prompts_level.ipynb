### Objective

Learn how prompt quality affects the quality of Large Language Model (LLM) responses by writing five different prompts for the same task, comparing the outputs, and analyzing which prompt performs best.

### What is a Prompt?

A prompt is the instruction or question given to an AI model to perform a task. The quality of the prompt directly affects the quality of the response.

### Why Prompt Engineering?

Prompt engineering helps produce:

More accurate responses
Better structured outputs
More relevant information
Fewer misunderstandings

Improving a prompt by adding context, a role, formatting requirements, and constraints usually results in higher-quality answers.

### Prompt Engineering Techniques Used

Throughout this exercise, the prompts gradually incorporate several prompt engineering techniques:

| Technique          | Description                                |
| ------------------ | ------------------------------------------ |
| Basic Prompt       | Simple instruction with little context     |
| Give a Role        | Assign the AI a profession or expertise    |
| Provide Details    | Add context about the task                 |
| Ask Step-by-Step   | Request organized reasoning or planning    |
| Specify Format     | Define how the output should be presented  |
| Add Constraints    | Limit length, number of items, style, etc. |
| Creative Prompting | Encourage more original ideas              |


### Selected Task
**Generate AI Engineering portfolio project ideas for a beginner AI engineer.**

### Prompt 1 — Basic 
A simple prompt that only states the task without providing any additional context, role, or formatting instructions. 
It serves as the baseline for comparing later prompts.

```python
prompt = """ Generate AI engineering portfolio project ideas. """
```

```python
from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    options = {
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 40,
            "repeat_penalty": 1.1,
            "num_predict": 256,
            "num_ctx": 1024,
            "seed": 42
        }
)

print(response.message.content)
```

### Prompt 2 — Improved
This prompt introduces a specific role for the AI, encouraging it to respond from the perspective of an experienced AI engineering mentor and produce more relevant recommendations.

```python
prompt = """You are an experienced AI engineering mentor.
            Generate AI engineering portfolio project ideas.
            """
```

```python
from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    options = {
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 40,
            "repeat_penalty": 1.1,
            "num_predict": 256,
            "num_ctx": 1024,
            "seed": 42
        }
)

print(response.message.content)
```

### Prompt 3 — Detailed
This prompt adds context about the user's goal and specifies the desired output format, resulting in more organized and tailored project ideas.

```python
prompt = """You are an experienced AI engineering mentor. 
            Generate AI engineering portfolio project ideas 
            for someone who has completed the fundamentals of AI and machine learning 
            and wants to build a strong GitHub portfolio. 
            For each project, include the project title, a brief description, the technologies to use, and the skills it demonstrates.
            """
```

```python
from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    options = {
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 40,
            "repeat_penalty": 1.1,
            "num_predict": 256,
            "num_ctx": 1024,
            "seed": 42
        }
)

print(response.message.content)
```

### Prompt 4 — Creative
This prompt encourages the AI to think creatively by requesting unique and innovative portfolio projects that stand out from common AI applications.

```python
prompt = """You are an experienced AI engineering mentor. 
            Generate 10 creative and unique AI engineering portfolio project ideas 
            that stand out from common chatbot and image classification projects. 
            Focus on real-world problems, emerging AI technologies, and projects that would impress recruiters. 
            For each project, include a catchy title, a short description, the main technologies to use, 
            and why it would be valuable in a portfolio.
            """
```

```python
from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    options = {
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 40,
            "repeat_penalty": 1.1,
            "num_predict": 256,
            "num_ctx": 1024,
            "seed": 42
        }
)

print(response.message.content)
```

### Prompt 5 — With Clear Constraints
This prompt combines role, context, formatting, and explicit constraints to generate well-structured, practical, and consistent project ideas that meet specific requirements.

```python
prompt = """You are an experienced AI engineering mentor helping a student build an outstanding GitHub portfolio. 
            Generate exactly 10 AI engineering portfolio project ideas. 
            Each project must be beginner-friendly but impressive enough for internships.

            Avoid generic ideas such as simple chatbots, calculators, or to-do apps. For each project, provide:
                1. Project title
                2. Problem it solves
                3. Brief description (2–3 sentences)
                4. Recommended technologies
                5. Difficulty level (Beginner, Intermediate, or Advanced)
                6. Estimated completion time
                7. Skills demonstrated

            Present the results as a Markdown table.
            """
```

```python
from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ],
    options = {
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 40,
            "repeat_penalty": 1.1,
            "num_predict": 256,
            "num_ctx": 1024,
            "seed": 42
        }
)

print(response.message.content)
```

### Comparison

| Prompt   | Techniques Used         | Output Quality | Observations                                               |
| -------- | ----------------------- | -------------- | ---------------------------------------------------------- |
| Prompt 1 | Task only               | Basic          | Generic ideas with little structure.                       |
| Prompt 2 | + Role                    | Better         | More relevant recommendations from the mentor perspective. |
| Prompt 3 | + Role + Context + Format | Very Good      | Better organized and tailored to portfolio building.       |
| Prompt 4 | + Creativity              | Excellent      | More original and recruiter-focused ideas.                 |
| Prompt 5 | + Constraints + Format    | Best           | Highly structured, actionable, and consistent output.      |


#### Which prompt gave the best result? and Why?
Prompt 5 produced the best result.

Prompt 5 clearly defined the AI's role, provided detailed context, specified the desired output format, and included explicit constraints. These additions reduced ambiguity and resulted in well-structured, practical, and consistent portfolio project ideas that were easier to understand and compare.

### Conclusion

In this exercise demonstrated that prompt quality has a significant impact on the quality of AI-generated responses. As more context, structure, and constraints were added, the outputs became increasingly relevant, organized, and useful. Prompt engineering is an effective way to guide large language models toward producing higher-quality results.
